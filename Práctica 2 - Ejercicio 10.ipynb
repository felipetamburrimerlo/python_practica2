## Apertura de Archivos

Primero abrimos los archivos en sus variables correspondientes:

```python
nombres = open('nombres_1.txt', encoding='UTF-8')
notas1 = open('eval1.txt')
notas2 = open('eval2.txt')
```

*Aclaraciones:*
* *No se especifica el modo de apertura de los archivos ya que por defecto está establecido **'r'**.*
* *Los archivos se deben encotrar en la misma carpeta que el código.*
* *El archivo **nombres_1.txt** se abre con el encoding **UTF-8** para que pueda reconocer correctamente todos los caracteres de los nombres.*

## Inicialización de Variables

Ahora inicializamos un diccionario vacío en el cuál se van a guardar los nombres y sus notas, así como también la variable *prom* en la que se calculará el promedio general:

```python
notas_totales = {}
prom = 0
```

*Aclaración: Si bien con el archivo **nombres_1.txt** se produce un error ya que hay nombres repetidos (por lo que al agregarlos al diccionario, varios nombres se reemplazan), fue visto en teoría que un diccionario es la manera correcta de organizar datos con valores asociados a los mismos, por esto es que lo que habría que hacer para que el programa funcione correctamente sería usar una clave unívoca (como el DNI o el legajo de los alumnos) como clave del diccionario en vez del primer nombre.*

## Carga del Diccionario

En este punto, usamos un *for* combinado con un *zip*, para poder iterar sobre los tres archivos a la vez en tres variables distintas. A cada valor de *nombre* se lo establece como clave del diccionario y a la suma de los valores de *nota1* y *nota2* (interpretados como integers) la establece como valor de dicha clave:

```python
for nombre, nota1, nota2 in zip(nombres, notas1, notas2):
    notas_totales[nombre.strip('\n,\' ')] = (
        int(nota1.strip('\n, ')) + int(nota2.strip('\n, ')))
```

*Aclaración: Debido a que en los archivos de texto los elementos están separados por **\n**, coma y espacio, así como también por comillas en el caso de los nombres, para obtener los elementos sin todos estos separadores, usamos la función **strip**, que remueve de una cadena de caracteres los caracteres recibidos como parámetros.*

## Cierre de los archivos

Debido a que los archivos ya no deben ser utilizados, se cierran:

```python
nombres.close()
notas1.close()
notas2.close()
```

## Cálculo del promedio

Para esto, recorremos el diccionario, sumando a la variable *prom* los valores (notas totales) de todos los nombres. Luego, redefinimos a la variable *prom* como la división de ella misma entre el tamaño de *notas_totales* (la cantidad de alumnos):

```python
for nombre in notas_totales:
    prom += notas_totales[nombre]

prom = prom / len(notas_totales)
```

*Aclaración: Si bien la suma de todas las notas podría hacerse mientras estas se cargan en el diccionario para evitar recorrer la estructura dos veces, creo que es más simple a la vista y fácil de entender si se recorre nuevamente en la sección del cálculo del promedio.*

## Informe de resultados

Finalmente, el programa debe imprimir una lista con los nombres de los alumnos cuya nota sea menor que el promedio. Para esto, primero imprime el encabezado de la lista y luego itera sobre la misma; para cada valor de *alumno*, si el valor asociado a esa clave (la nota del mismo) es menor que el promedio, se imprime el nombre:

```python
print('Alumnos que tuvieron menor nota que el promedio:')

for alumno in notas_totales:
    if notas_totales[alumno] < prom:
        print(alumno)
```
